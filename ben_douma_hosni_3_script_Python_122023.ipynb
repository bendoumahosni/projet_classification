# recuperation de resultats

```python

if response.status_code == 200:
    #  JSON => DataFrame pandas
    data = response.json()
    df = pd.json_normalize(data['hints'])  
    print(df)
else:
    print(f"Erreur de requête: {response.status_code}")
    # affichage  texte de la reponse
    print(response.text)
```
